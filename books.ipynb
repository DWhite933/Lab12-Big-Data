```python
# create cursor
cursor = db.cursor()
```

```python
# A
pd.read_sql(""" SELECT last
                FROM authors 
                ORDER BY last DESC""", db)
```

```python
# B
pd.read_sql(""" SELECT title
                FROM titles
                ORDER BY title ASC""", db)
```

```python
# C 
pd.read_sql("""SELECT id, title, copyright 
                     FROM titles
                     INNER JOIN author_ISBN
                        ON titles.isbn = author_ISBN.isbn
                     WHERE id = 1
                     ORDER BY title ASC""", db)
```

```python
# D
# add author to authors
cursor = cursor.execute("""INSERT INTO authors (first, last)
                           VALUES ('Al','Sweigart')""")

pd.read_sql("SELECT id, first, last FROM authors", db, index_col=['id'])
```

```python
# E
# add book to titles
cursor = cursor.execute("""INSERT INTO titles (isbn, title, edition, copyright)
                           VALUES (1593279922, 'Automate the Boring Stuff', 2, 2019)""")

pd.read_sql("SELECT * FROM titles", db)
```
